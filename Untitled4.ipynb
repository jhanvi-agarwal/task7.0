```python
import sqlite3

# Step 1: Create and connect to SQLite database
conn = sqlite3.connect("sales_data.db")
cursor = conn.cursor()

# Step 2: Create table
cursor.execute("""
CREATE TABLE IF NOT EXISTS sales (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    product TEXT NOT NULL,
    quantity INTEGER NOT NULL,
    price REAL NOT NULL
)
""")

# Step 3: Insert dummy data
sales_data = [
    ("Laptop", 3, 50000),
    ("Laptop", 2, 52000),
    ("Mobile", 5, 15000),
    ("Mobile", 8, 15500),
    ("Tablet", 4, 30000),
    ("Tablet", 6, 31000),
    ("Headphones", 10, 2000),
    ("Headphones", 15, 2200),
    ("Smartwatch", 7, 5000),
    ("Smartwatch", 9, 5500)
]

cursor.executemany("INSERT INTO sales (product, quantity, price) VALUES (?, ?, ?)", sales_data)

# Step 4: Commit changes and close connection
conn.commit()
conn.close()

print("Database 'sales_data.db' created with dummy sales data!")
```

```python
# Step 1: Import libraries
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Step 2: Connect to SQLite database
# Agar tumhare paas 'sales_data.db' file same folder me hai toh direct connect ho jayega
conn = sqlite3.connect("sales_data.db")

# Step 3: SQL query to get total quantity & total revenue per product
query = """
SELECT 
    product, 
    SUM(quantity) AS total_qty, 
    SUM(quantity * price) AS revenue
FROM sales
GROUP BY product
"""

# Step 4: Load query result into pandas DataFrame
df = pd.read_sql_query(query, conn)

# Step 5: Print the results
print("Sales Summary:")
print(df)

# Step 6: Plot bar chart for revenue per product
plt.figure(figsize=(8,5))
df.plot(kind='bar', x='product', y='revenue', legend=False)
plt.title("Revenue per Product")
plt.ylabel("Revenue")
plt.xlabel("Product")
plt.xticks(rotation=45)
plt.tight_layout()

# Step 7: Save the chart as PNG
plt.savefig("sales_chart.png")

# Step 8: Show the chart
plt.show()

# Step 9: Close the connection
conn.close()
```
